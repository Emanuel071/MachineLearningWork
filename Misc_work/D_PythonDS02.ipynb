## Manipulating Numpy Arrays

```python
m1da = np.zeros(10)
print(m1da)

m1da[0] = 50
print(m1da)

m1da[3:6] = 50
print(m1da)
```

```python
print(np.where(m1da == 50))

m2da = np.array([[1,5,9],[8,5,5]])
print(m2da)

print(np.where(m2da == 5))
print(np.where(m2da < 5))
print(np.where(m2da >= 5))
print(np.where(m2da != 5))
```

```python
print(np.argwhere(m2da==5))

index = np.where(m2da >= 5)
print(index)

print(m2da[index])

m2da[index] = 100
print(m2da)
```

```python
print(np.all(m1da))
print(np.all(m1da>=0))
print(np.all(m1da>5))
```

```python
print(np.any(m1da))

a = np.array([[1,2],[3,4]])
b = np.array([[5,6],[7,8]])
print(a)
print(b)

v = np.vstack((a,b))
print(v)

print(np.vstack((a,b,a,b)))

h = np.hstack((a,b))
print(h)
print(np.hstack((a,b,a,b)))
```

## ** MINI PROJECT ** Image Manipulation using Numpy

```python
from skimage import io
import matplotlib.pyplot as plt

```

```python
data_path = "C:/Users/eacalder/Documents/Education/DataScienceInfinity/Leasson_Downloads/D_PythonDS02/camaro.jpg"
```

```python
camaro = io.imread(data_path)
print(camaro)
```

```python
camaro.shape #wider than it is tall 
```

```python
plt.imshow(camaro)
plt.show()
```

```python
# cropping image vertically
cropped = camaro[0:500,:,:]
plt.imshow(cropped)
plt.show()
```

```python
# cropping image horzontal   
hcropped = camaro[:,400:1000,:]
plt.imshow(hcropped)
plt.show()
```

```python
# cropping image 
cropped = camaro[350:1100,200:1400,:]
plt.imshow(cropped)
plt.show()
```

```python
io.imsave('camaro_cropped.jpg', cropped)
```

```python
#flip image
vflipped = camaro[::-1,:,:]
plt.imshow(vflipped)
plt.show()
```

```python
#flip image
hflipped = camaro[:,::-1,:]
plt.imshow(hflipped)
plt.show()
```

```python
# color channels
red = np.zeros(camaro.shape, dtype='uint8')
# creating an array with exact dimesions of the camaro only filled with zeros
print(red)
```

```python
red[:,:,0] = camaro[:,:,0]
plt.imshow(red)
plt.show()
```

```python
green = np.zeros(camaro.shape, dtype='uint8')
green[:,:,1] = camaro[:,:,1]
plt.imshow(green)
plt.show()
```

```python
blue = np.zeros(camaro.shape, dtype='uint8')
blue[:,:,2] = camaro[:,:,2]
plt.imshow(blue)
plt.show()
```

```python
camaro_rainbow = np.vstack((red,green,blue))
plt.imshow(camaro_rainbow)
plt.show()
```
