```python
########################
#Association rule learning
########################

########################
#Import required packages
########################
from apyori import apriori
import pandas as pd
```

```python
########################
#Import data
########################

#import 
alcohol_transactions = pd.read_csv('Saved_files/sample_data_apriori.csv')
print(alcohol_transactions.shape)
print(alcohol_transactions.info())
alcohol_transactions.head()
```

```python
# drop id column
transaction_id = alcohol_transactions['transaction_id']
alcohol_transactions.drop('transaction_id', axis=1, inplace=True)
print(alcohol_transactions.shape)
alcohol_transactions.head()
```

```python
# modify data for apriori algorithm
# doesnt accept in the form of pandas dataframe please provide as a list of list
# we have to loop and append assuring those that are blank not in list
transaction_list = []
for index, row in alcohol_transactions.iterrows():
    transaction = list(row.dropna())
    transaction_list.append(transaction)
print(transaction_list)
```

```python
########################
# apply apriori algorithm
########################
# after need to convert to a list as it is in the generator form
apriori_rules = apriori(transaction_list,
                        min_support = 0.003,# just used this as it went well but test out
                        min_confidence = 0.2,# of all transaction item a included b
                        min_lift = 3,
                        min_length = 2,
                        max_length = 2 
                        )

apriori_rules = list(apriori_rules)

#132 rules that met the criteria 
len(apriori_rules)
```

```python
apriori_rules[0]
#apriori_rules[0][2][0][0]
```

```python
########################
#convert output to dataframe
########################
product1 = [list(rule[2][0][0])[0] for rule in apriori_rules]
print(product1)
product2 = [list(rule[2][0][1])[0] for rule in apriori_rules]
support = [rule[1] for rule in apriori_rules]
confidence = [rule[2][0][2] for rule in apriori_rules]
lift = [rule[2][0][3] for rule in apriori_rules]
```

```python
apriori_rules_df = pd.DataFrame({'product1':product1,
                                 'product2':product2,
                                 'support':support,
                                 'confidence':confidence,
                                 'lift':lift})

apriori_rules_df.head()
```

```python
########################
#sort rules by descending lift
########################
apriori_rules_df.sort_values(by='lift', 
                             ascending=False, 
                             inplace=True)
apriori_rules_df.head(10)
```
